```python
######## SQL connection ############

import pypyodbc # pip install pypyodbc
import pandas as pd # pip install pandas

SERVER_NAME = 'DESKTOP-H2M6ONO'
DATABASE_NAME = 'test'

conn = pypyodbc.connect("""
    Driver={{SQL Server Native Client 11.0}};
    Server={0};
    Database={1};
    Trusted_Connection=yes;""".format(SERVER_NAME, DATABASE_NAME)
)

sql_query1 = """
SELECT * FROM organization_provider WITH (NoLock)
"""

# With Headers
organization_provider = pd.read_sql(sql_query1, conn)


sql_query2 = """
SELECT * FROM Encounters_medical_details WITH (NoLock)
"""

# With Headers
Encounters_medical_details = pd.read_sql(sql_query2, conn)

sql_query3 = """
SELECT * FROM patients_filtered WITH (NoLock)
"""

# With Headers
patients_filtered = pd.read_sql(sql_query3, conn)
```
